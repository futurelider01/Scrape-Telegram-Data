```python
from deep_translator import GoogleTranslator
import pandas as pd

df = pd.read_csv('tg_msg.csv', header='infer')
```

```python
def tranlate(text):
    tr = GoogleTranslator(target='en').translate(text)
    return tr
```

```python
df = df[df['ID']<44940]
```

```python
for i, df_ in enumerate(df.values):
    row = f'{df_[0]},{tranlate(df_[1])},{df_[2]}\n'
    with open('tg_msg_en_new.csv','a',encoding='utf-8') as file:
        file.write(row)
    print(f"{i+1}-row is proceeded")
```
